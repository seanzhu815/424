```python
consumption_growth
```

```python
gamma = 2

# Calculate the covariance between asset returns and consumption growth
covariance = np.cov(returns, consumption_growth)[0,1]

# Calculate the risk premium
risk_premium = gamma * covariance

print("Risk Premium:", risk_premium)
```
